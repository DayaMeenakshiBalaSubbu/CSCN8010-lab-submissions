```python
#Part 1 (5 points)

# 1. Get the data.
# Importing the required libraries and loading the diabetes dataset
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as pt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error


#Extracting the dataset as a Pandas and displaying the data 
diabetes = datasets.load_diabetes(as_frame=True)
diabetes_df = diabetes.frame  
diabetes_df.head()
```

```python
# 2. Frame the Problem. Notice that the dependent variable of diabetes progression, 
# namely "disease progression one year after baseline".

#Displaying the column names from the diabetes dataframe
print("Column Names:", diabetes_df.columns)

#The variable x indicates the features and y shows the target variable
X=diabetes_df.drop(columns=['target']) 
y=diabetes_df['target']
# Displaying the the shape and name of the variables with the independent variables and dependent variable.
print(f"Features Shape: {X.shape}, Target Shape: {y.shape}")
print(f"Independent Variables from the dataset: {list(X.columns)}")
print(f"Dependent Variable from the dataset: 'target'")
print("Observation:")
print("The target variable is namely disease progression one year after baseline which is target column.")
print("The given diabetes dataset has ten features such as'age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6','target' ")
print("The goal is to predict the diabetes progression using the different models")
  
```

```python
#3. EDA - Describe the data, explore it, and provide insights about it. 
# This should include at least: statistics, scatter plots histograms, a correlation matrix, and concise and relevant insights (4 point).

# Display basic statistics
print("Statistics Data:", diabetes_df.describe())

#Scatter Plots Histogram for all the features
print("Scatter Plot")
diabetes_df.hist(figsize=(10, 10), bins=20)
pt.show()
#Scatter plot for BMI and target
pt.figure(figsize=(8, 4))
sns.scatterplot(x=diabetes_df['bmi'], y=diabetes_df['target'])
pt.title("Scatter plot for BMI vs Target (Disease Progresssion)")
pt.xlabel("BMI")
pt.ylabel("Target (Disease Progresssion)")
pt.show()

#Correlation matrix
pt.figure(figsize=(8, 8))
sns.heatmap(diabetes_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
pt.title("Correlation Matrix")
pt.show()

#Summary and Insights
print("Insights:")
print("Correlation is stronger between the BMI and target from the dataset.")
print("The datasets is already cleaned and there is no missing values in it. So, normalization is not required for this dataset.")
```

```python
#4. Clean the data if needed, and explain your reasoning for your reader (1 point)

# Checking for missing values for data cleaning process
print(diabetes_df.isnull().sum())
print("There is no missing value in the dataset for all the features. So data cleaning is not required for this dataset.")

```

```python
#5. Split the dataset to a train (75%) and validation set (10%), and test set (15%).

# Splitting the dataset to a 75% Train, 10% Validation, 15% Test set
train_X, temp_X, train_y, temp_y = train_test_split(X, y, test_size=0.25, random_state=42)
Validation_X, test_X, Validation_y, y_test = train_test_split(temp_X, temp_y, test_size=0.6, random_state=42)

# Displaying the final splitted dataset values
print(f"Train set: ",train_X.shape)
print ("Train Dataset first five values: \n", train_X.head())

print(f"Validation set: ",Validation_X.shape)
print ("Validation Dataset first five Values: \n", Validation_X.head())

print(f"Test set: ",test_X.shape)
print ("Test Dataset first five values: \n", test_X.head())
```

```python
#Part 2
#6. Models: a univariate polynomial regression on the BMI feature versus the "disease progression one year after baseline" 
# - from degree 0 to 5 (6 models) (1.5 point)

#Importing the required libraries
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

#Selecting BMI and target variable to fit the models from degree 0 to 5
train_X_bmi = train_X[['bmi']]
Validation_X_bmi = Validation_X[['bmi']]
test_X_bmi = test_X[['bmi']]
train_y = train_y
Validation_y = Validation_y
y_test = y_test

#Dictionary to store the models
m = {}

#Fit the polynomial regression models
for degree in range(6):
    polynomial = PolynomialFeatures(degree=degree)
    train_X_polynomial = polynomial.fit_transform(train_X_bmi)
    Validation_X_polynomial = polynomial.transform(Validation_X_bmi)
    
    #Train and store the model and polynomial transformation
    p_model = LinearRegression()
    p_model.fit(train_X_polynomial, train_y)
    
    #Store the model and polynomial transformation
    m[degree] = (p_model, polynomial)

    #Displaying the coefficients of the trained model

    print(f"Degree {degree}: ")
    print(f"Coefficients {p_model.coef_}")
    print(f"Intercept {p_model.intercept_}")
    print(f"")
```

```python
#7. Compare the models using the training and validation data. Construct a table summarizing the train validation results. Each model should have a separate row in the table (3 points):
#    1. R-Squared
#    2. Mean Absolute Error (MAE)
#    3. MAPE

# Function to calculate Mean Absolute Percentage Error (MAPE)
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Dictionary to store output
output = []

# Evaluate each polynomial model
for degree, (p_model, polynomial) in m.items():
    #Transforming the data using the polynomial features
    train_X_polynomial = polynomial.transform(train_X_bmi)
    Validation_X_polynomial = polynomial.transform(Validation_X_bmi)
    # Make predictions and evaluating the metrics and storing them in a list
    train_pred = p_model.predict(train_X_polynomial)
    Validation_pred = p_model.predict(Validation_X_polynomial)
    train_r2 = r2_score(train_y, train_pred)
    Validation_r2 = r2_score(Validation_y, Validation_pred)
    train_mae = mean_absolute_error(train_y, train_pred)
    Validation_mae = mean_absolute_error(Validation_y, Validation_pred)
    train_mape = mean_absolute_percentage_error(train_y, train_pred)
    Validation_mape = mean_absolute_percentage_error(Validation_y, Validation_pred)
    output.append([degree, train_r2, Validation_r2, train_mae, Validation_mae, train_mape, Validation_mape])

# Converting the output into a dataframe
final_df = pd.DataFrame(output, columns=['Degree', 'Train R2', 'Validation R2', 'Train MAE', 'Validation MAE', 'Train MAPE', 'Validation MAPE'])
# Displaying the final dataframe output

print("Trained Dataset for R-Squared, Mean Absolute Error (MAE),  MAPE : ")
print(final_df)
```

```python
#8. Identify the best model based on the table above.
# Identify the best model based on validation R² and MAE (lowest MAE, highest R²)
best_degree = final_df.loc[final_df['Validation R2'].idxmax(), 'Degree']
best_validation_r2 = final_df.loc[final_df['Validation R2'].idxmax(), 'Validation R2']
best_validation_mae = final_df.loc[final_df['Validation R2'].idxmax(), 'Validation MAE']
print(f"Best Model between the range 0 to 5: Degree {best_degree}")
print(f"Validation R²: {best_validation_r2}")
print(f"Validation MAE: {best_validation_mae}")

print(f"The best model based on validation R2 and MAE is degree 5. The validation R2 is higher and validation MAE is lower.")
```

```python
#9. Run the chosen model on the test set and provide results (R-Squared, MAPE, MAE) (1 point).

#Getting the best model and transform the test set using the polynomial features.
best_model, best_polynomial = m[5]
test_X_polynomial = best_polynomial.transform(test_X_bmi)

#Predict the target variable on the test set and evaluate the R2, MAE and MAPE for the test set.
test_predict = best_model.predict(test_X_polynomial)
test_r2 = r2_score(y_test, test_predict)
test_mae = mean_absolute_error(y_test, test_predict)
test_mape = mean_absolute_percentage_error(y_test, test_predict)

# Display the results
print(f"Test Set output for Best Model from the above Degree 5 :")
print(f"R-Squared:",test_r2)
print(f"Mean Absolute Error (MAE):",test_mae)
print(f"Mean Absolute Percentage Error (MAPE): {test_mape:2f}%")
```

```python
#10. Plot a graph of the train, validation and test data points, and the fit of the chosen model over each of them (1 point).

#Plotting the train data, validation data and test data points
pt.figure(figsize=(10, 6))
pt.scatter(train_X_bmi, train_y, color='blue', label='Train Data', alpha=0.6)
pt.scatter(Validation_X_bmi, Validation_y, color='green', label='Validation Data', alpha=0.6)
pt.scatter(test_X_bmi, y_test, color='brown', label='Test Data', alpha=0.6)

#Plot the polynomial fit over the range of minimum to maximum of BMI Value
bmi_range = np.linspace(min(train_X_bmi['bmi']), max(train_X_bmi['bmi']), 100).reshape(-1, 1)
bmi_range_polynomial = best_polynomial.transform(bmi_range)
pred = best_model.predict(bmi_range_polynomial)
#Plotting the lines and labeling the plots.
pt.plot(bmi_range, pred, color='red', label='Polynomial Degree 5', linewidth=2)
pt.title('Train, Validation, Test Data Points and Fit for Polynomial Regression Model with the Degree 5')
pt.xlabel('BMI')
pt.ylabel('Disease Progression (Target)')
pt.legend(loc='upper left')
pt.show()
```

```python
#11. Write down the equation of the best model (with a two decimal digit precision, assuming it's sufficient) (1 point).

# Getting the coefficients and intercept and creatng the polynomial equation.
cf = np.round(best_model.coef_, 2)  # Round to 2 decimal places
intercept = round(best_model.intercept_, 2)  # Round to 2 decimal places

# Create the polynomial equation
val = [f"{coeff} * BMI^{i}" for i, coeff in enumerate(cf)]
equation = " + ".join(val)
equation = f"{intercept} + {equation}"

# Print the equation
print("Equation of the best polynomial model:")
print(equation)
```

```python
#13. How many trainable parameters are we fitting for each of the models? Explain these values. One way is to use sklearn function `get_feature_names_out()`(1 point)

# Loop through each polynomial degree (0 to 5)
for degree in range(6):
    # Create polynomial feature transformer
    polynomial = PolynomialFeatures(degree=degree)
    # Transform training BMI feature to get feature names
    polynomial.fit(train_X_bmi)
    
    # Get feature names (trainable parameters)
    n = polynomial.get_feature_names_out(['bmi'])
    
    # Number of trainable parameters = Number of coefficients in the model
    total_parameters = len(n)  # Intercept + coefficients
    print(f"Degree",degree)
    print(f"Trainable Parameters = {total_parameters}, Features: {n}")
```

```python
#2. Two decision trees (e.g. consider changing the `max_depth` hyperparameter)

#Importing Decision Tree library
from sklearn.tree import DecisionTreeRegressor

#Defining the tree depths and storing the model into the dictionary.
tree_depths = [3, 5]
tree_models = {}
#Loop through each depth and train the decision tree model and make predictions.
for depth in tree_depths:
    #Train Decision Tree model
    tree = DecisionTreeRegressor(max_depth=depth, random_state=42)
    tree.fit(train_X, train_y)
    train_preds = tree.predict(train_X)
    validation_preds = tree.predict(Validation_X)
    
    #Calculate metrics
    train_r2=r2_score(train_y, train_preds)
    validation_r2=r2_score(Validation_y, validation_preds)
    validation_mae=mean_absolute_error(Validation_y, validation_preds)
    validation_mape=mean_absolute_percentage_error(Validation_y, validation_preds)
    
    #Store and print output
    tree_models[depth] = {
        'model': tree,
        'train R2': train_r2,
        'validation R2': validation_r2,
        'validation MAE': validation_mae,
        'validation MAPE': validation_mape
    }
    print("-" * 50)
    print(f"Decision Tree (max_depth={depth}):")
    print(f"Train R2: {train_r2:.4f}, Validation R2: {validation_r2:.4f}")
    print(f"Validation MAE: {validation_mae:.4f}, Validation MAPE: {validation_mape:.4f}")
    print("-" * 50)
```
